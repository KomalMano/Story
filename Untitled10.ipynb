```python
import pandas as pd
from google.colab import files
import io
#now here is all about excel file
```

```python
uploaded = files.upload()
```

```python
for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```

```python
uploaded
```

```python
df = pd.read_excel(io.BytesIO(uploaded['stock_data.xlsx']))
df
#now u see that we create dataFrame successfully
```

```python
#now u want to do conversion of cell content from ur cell 
def convert_people_cell(cell):
  if cell=="n.a.":
    return 'sam walton'
  return cell

def convert_people_cell(cell):
  if cell=="not avaiable":
    return None
  return cell

df = pd.read_excel(io.BytesIO(uploaded['stock_data.xlsx']),converters = {
    'people' : convert_people_cell,
    'eps' : convert_people_cell
})
df
```

```python
#last thing i want to writing my files
df.to_excel("new.xlsx",sheet_name="stocks")
```

```python
#we want to take the space in the excel file
df.to_excel("new.xlsx",sheet_name="stocks", startrow=1, startcol=2)
```

```python
#we want to take the excel file as e tiz
df.to_excel("new.xlsx",sheet_name="stocks", startrow=1, startcol=2, index=0)
```
