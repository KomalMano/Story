```python
#pyplot is the main plot object that u get in main plot lib module 
import matplotlib.pyplot as plt
#whe also using matplotlib inline magic
%matplotlib inline
```

```python
#here i just created weather data list
x=[1,2,3,4,5,6,7]#x is num of week days
y=[50,51,52,48,47,49,46]#y has temp
```

```python
#now simply plot this
plt.plot(x,y)
#now that will plot a 2D chart
```

```python
#the APi of plot 
#i just google it for api of the plot
#u just also apply the line thiker 
#u also just create the linestyle dashdot and dotted
plt.plot(x,y, color='green', linewidth=5, linestyle='dotted' )
```

```python
#now labelling x and y sides
plt.xlabel('Day')
plt.ylabel('Temprature')
plt.title('Weather') #also set my title chart
plt.plot(x,y, color='green', linewidth=5, linestyle='dotted' )
```
