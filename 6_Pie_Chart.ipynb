```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals=[1400,600,300,410,250]
exp_labels=["Home Rent","Food","Phone/Internet Bill","Car","Other Utilities"]
```

```python
#in the 1st argument supply expense values and there is keyword as an argument called labels where u supply the labels
plt.pie(exp_vals, labels=exp_labels)
```

```python
#for creating equally circle shape
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels)
```

```python
#i dont want to show all the text
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels)
plt.show()
```

```python
#we want to increase the size of the pie chart
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5) #radius default value is 1
plt.show()
```

```python
#now use the %age for the values use apply
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.f%%') #autopct use for %age values
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.1f%%') #autopct use for %age values but depends on the decimal points
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.1f%%', shadow=True) #for shadow
plt.show()
```

```python
#now take out these pies
plt.axis("equal")
#and the explode use arrays having number of pies involve in it. and use float values for seprate them.
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.1f%%', shadow=True, explode=[0,0.2,0.2,0,0]) #for explode use for split these pies
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.1f%%', shadow=True, explode=[0,0.2,0.2,0,0], startangle=45) #startangle use for rotate the pie angle position
plt.show()
```
