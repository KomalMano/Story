```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#use static data for day and temp
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y, 'g+')#g means green and plus means marker
```

```python
#for dashing
plt.plot(x,y, 'g+--')
```

```python
#u can make it dash dot also
plt.plot(x,y, 'g+-.')
```

```python
#so the oredr does not matter and it has the same effect
plt.plot(x,y, '--r*')
```

```python
#d fro diamond and r for red
plt.plot(x,y, 'rD--')
```

```python
plt.plot(x,y, 'rD')
```

```python
#u can achieve for individual argument working is same
plt.plot(x,y, color='red', marker='D', linestyle='')
```

```python
#u can also use hexadecimal numbers in color arguments
plt.plot(x,y, color='#008040', marker='D', linestyle='')
```

```python
#if u want to controle the marker size argument now its really big
plt.plot(x,y, color='#888888', marker='+', linestyle='', markersize=17)
```

```python
#alpha property of plot API.
#alpha will basically control transparancy
#so the alpha scale is basically b/w 0 1
#when alpha value is 1 it is opaque 
plt.plot(x,y, color='green', alpha=1)
```

```python
#zero is for visible
plt.plot(x,y, color='green', alpha=0.2)
```
