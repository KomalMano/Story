```python
#i got day as day time
df = pd.read_csv(io.StringIO(uploaded['new_weather_data.csv'].decode('utf-8')), parse_dates=["day"])
df
```

```python
#i want to make this day column in my index as my dataframe 
df = pd.read_csv(io.StringIO(uploaded['new_weather_data.csv'].decode('utf-8')), parse_dates=["day"])
df.set_index('day',inplace=True)#remember that inplace = true otherwise it will never modify the orignal dataframe
df
```

```python
#now i want to replace my NaN values into an other values.
#now here is a 1st method we are going to cover is fillna
new_df = df.fillna(0)
new_df
```

```python
#now fillna is best for temp and wind but not best for event so in this case pass ditionary
new_df = df.fillna({
    'temp':0,
    'windspeed':'0',
    'event':'no event'
})
new_df
```

```python
#when u calculate mean then its zero value is not good.So the previous value carry forward to the next value.
new_df = df.fillna(method="ffill")
new_df
```

```python
#backward value copy
new_df = df.fillna(method="bfill")
new_df
```

```python
#vertically copy data backward side
new_df = df.fillna(method="bfill", axis="columns")
new_df
```

```python
#forward fill have limit 1 which is only copy 1 time only
new_df = df.fillna(method="ffill", limit=1)
new_df
```

```python
#forward fill have limit 2 which is only copy 2 time only
new_df = df.fillna(method="ffill", limit=2)
new_df
```

```python
#i dont no the accurate temp.So the temp is in b/w the exact values.upper and lower.
#for interpolate use graduatually upas and downs
new_df = df.interpolate()
new_df
```

```python
#here somtimes data is missing just like date and the temp is not accurate interpolate bcz days missing. So use time to create accurate value.
new_df = df.interpolate(method="time")
new_df 
```

```python
#if i have missing values so drop all these rows
new_df = df.dropna()
new_df
```

```python
#if i want to drop only 1 row which has at least 1 missing data
new_df = df.dropna(how='all')
new_df
```

```python
#if i have at least 1 NAN value then keep the row
new_df = df.dropna(thresh=1)#its same the previous 
new_df
```

```python
#if i have at least 2 NAN value then keep the row.Means 2 valid values
new_df = df.dropna(thresh=2) 
new_df
```

```python
#last thing is,How i inserting the missing date
dt = pd.date_range("01-01-2017","01-11-2017")
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
```
